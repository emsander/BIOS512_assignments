```r
wind_turbine |> head()
```

Each row in the table represents a single turbine.

Make a bar chart üìä **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
library('tidyverse')
```

```r
wind_turbine_provsort = wind_turbine |>
    mutate(province_territory_sorted = province_territory |> fct_infreq() |> fct_rev())
```

```r
options(repr.plot.width=4, repr.plot.height=4, repr.plot.res = 250)
p = ggplot(wind_turbine_provsort, aes(x = province_territory_sorted)) +
    geom_bar(fill = 'chartreuse4') +
    labs(x = "Province or Territory", y = "Number of Turbines") +
    coord_flip() + 
    theme_minimal() + 
    theme(axis.title.x = element_text(size = 8),
        axis.title.y = element_text(size = 8),
        axis.text.y = element_text(size = 6))

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
wind_turbine_manfsort = wind_turbine |>
    mutate(manufacturer_sorted = manufacturer |> 
           fct_lump(n = 7) |> 
           fct_infreq() |> 
           fct_relevel('Other', after= Inf) |> 
           fct_rev())
```

```r
options(repr.plot.width=4, repr.plot.height=4, repr.plot.res = 250)
p = ggplot(wind_turbine_manfsort, aes(x = manufacturer_sorted)) +
    geom_bar(fill = 'steelblue') +
    labs(x = "Manufacturer", y = "Number of Turbines") +
    coord_flip() + 
    theme_minimal() + 
    theme(axis.title.x = element_text(size = 10),
        axis.title.y = element_text(size = 10),
        axis.text.y = element_text(size = 8))

p
```
